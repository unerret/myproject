```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('/content/drive/My Drive/rideshare_2022_cleaned.csv')
clean =df.dropna()
clean.head(15)
```

```python
ac = clean['additional_charges'].head(15)
tt = clean['trip_total'].head(15)
print(ac)
print(tt)
```

```python
nac = np.array(ac)
ntt = np.array(tt)

t_test, p_value = stats.ttest_ind(nac, ntt,equal_var=True)
t_test2, p_value2 = stats.ttest_ind(nac, ntt)
print(t_test, p_value)
```

```python
df = 28
alpha = 0.05
t_critical = stats.t.ppf(1-alpha/2,df)
print(t_critical)
```

```python
if(t_critical < abs(t_test)):
  print("Reject null hypothesis")
else:
  print("Accept null hypothesis")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/drive/My Drive/rideshare_2022_cleaned.csv')
clean =df.dropna()
clean.head(15)
```

```python
x = clean[['additional_charges']]
y = clean['trip_total']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
x = clean[['additional_charges']]
y = clean['trip_total']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
print(f"the dataset for training is {xtrain}")
print(f"the dataset for testing is {xtest}")
print(f"the target for training is {ytrain}")
print(f"the target for testing is {ytest}")
```

```python
model = LinearRegression().fit(xtrain,ytrain)
```

```python
ypredict = model.predict(xtrain)
print (f"the predicted values are {ypredict}")
```

```python
ypredict = model.predict(xtest)
print (f"the independent values are {xtest}")
print (f"the dependent values are {ypredict}")
```
